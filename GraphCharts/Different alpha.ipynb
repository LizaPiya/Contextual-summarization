```python
import matplotlib.pyplot as plt

# Data
alpha_values = [0.2, 0.5, 0.7, 0.9]
bleu_1 = [9.90, 6.91, 9.86, 9.98]
bleu_2 = [3.34, 2.64, 3.24, 3.50]
rouge_l = [10.15, 9.81, 9.94, 10.08]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(6.5, 4.5))

# Plot lines with enhanced formatting
ax.plot(alpha_values, bleu_1, label="BLEU-1", marker="o", markersize=7, linestyle="-", linewidth=1.5, color="tab:blue")
ax.plot(alpha_values, bleu_2, label="BLEU-2", marker="s", markersize=7, linestyle="--", linewidth=1.5, color="tab:orange")
ax.plot(alpha_values, rouge_l, label="ROUGE-L", marker="^", markersize=7, linestyle=":", linewidth=1.5, color="tab:green")

# Labels and ticks
ax.set_xlabel(r"Weight Factor for Layer Importance ($\alpha$)", fontsize=13, fontweight="bold")
ax.set_ylabel("Evaluation Metric Score", fontsize=13, fontweight="bold")
ax.set_xticks(alpha_values)
ax.set_xticklabels([f"{a:.1f}" for a in alpha_values], fontsize=12)
ax.set_yticks(range(0, 15, 3))
ax.set_ylim(0, 15)

# Grid and border formatting
ax.grid(True, linestyle="--", linewidth=0.6, alpha=0.7)
ax.spines["top"].set_linewidth(1.2)
ax.spines["right"].set_linewidth(1.2)
ax.spines["left"].set_linewidth(1.2)
ax.spines["bottom"].set_linewidth(1.2)

# Legend formatting
ax.legend(fontsize=11, loc="upper left", frameon=True, edgecolor="black", fancybox=True)

# Optimize layout and save
plt.tight_layout()
plt.savefig("alpha_evaluation.pdf", format="pdf", dpi=300)

# Show plot
plt.show()
```
