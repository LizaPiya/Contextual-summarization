```python
import matplotlib.pyplot as plt

# Data
alpha_values = [0.2, 0.5, 0.7, 0.9]
bleu_1 = [9.90, 6.91, 9.86, 9.98]
bleu_2 = [3.34, 2.64, 3.24, 3.50]
rouge_l = [10.15, 9.81, 9.94, 10.08]

# Create the figure and axis
fig, ax = plt.subplots(figsize=(6.5, 4.5))

# Plot lines with enhanced formatting
ax.plot(alpha_values, bleu_1, label="BLEU-1", marker="o", markersize=7, linestyle="-", linewidth=1.5, color="tab:blue")
ax.plot(alpha_values, bleu_2, label="BLEU-2", marker="s", markersize=7, linestyle="--", linewidth=1.5, color="tab:orange")
ax.plot(alpha_values, rouge_l, label="ROUGE-L", marker="^", markersize=7, linestyle=":", linewidth=1.5, color="tab:green")

# Labels and ticks
ax.set_xlabel(r"Weight Factor for Layer Importance ($\alpha$)", fontsize=13, fontweight="bold")
ax.set_ylabel("Evaluation Metric Score", fontsize=13, fontweight="bold")
ax.set_xticks(alpha_values)
ax.set_xticklabels([f"{a:.1f}" for a in alpha_values], fontsize=12)
ax.set_yticks(range(0, 15, 3))
ax.set_ylim(0, 15)

# Grid and border formatting
ax.grid(True, linestyle="--", linewidth=0.6, alpha=0.7)
ax.spines["top"].set_linewidth(1.2)
ax.spines["right"].set_linewidth(1.2)
ax.spines["left"].set_linewidth(1.2)
ax.spines["bottom"].set_linewidth(1.2)

# Legend formatting
ax.legend(fontsize=11, loc="upper left", frameon=True, edgecolor="black", fancybox=True)

# Optimize layout and save
plt.tight_layout()
plt.savefig("alpha_evaluation.pdf", format="pdf", dpi=300)

# Show plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Data from the tables
models = ['Vanilla LLaMA 3.2', 'LLaMA 3.2 + CPTF']
metrics = ['BLEU-1', 'BLEU-2', 'ROUGE-L']

# Metrics for Vanilla LLaMA 3.2 (from the first table, max token = 300, temperature = 0.1)
vanilla_metrics = [6.91, 2.64, 9.81]

# Metrics for LLaMA 3.2 + CPTF (from the second table)
cptf_metrics = [10.08, 3.36, 9.62]

# Bar chart settings
x = np.arange(len(metrics))  # Label locations
width = 0.25  # Width of bars

# Colors for bars
vanilla_color = '#6baed6'  # Soft blue
cptf_color = '#74c476'  # Soft green

# Plotting
fig, ax = plt.subplots(figsize=(8, 5))
bars1 = ax.bar(x - width/2, vanilla_metrics, width, label='LLaMA 3.2', color=vanilla_color, edgecolor='black')
bars2 = ax.bar(x + width/2, cptf_metrics, width, label='LLaMA 3.2 + CPTF', color=cptf_color, edgecolor='black')

# Adding labels and title
ax.set_xlabel('Metrics', fontsize=12)
ax.set_ylabel('Scores (%)', fontsize=12)
#ax.set_title('Performance Metrics: LLaMA 3.2 vs LLaMA 3.2 + CPTF', fontsize=14)
ax.set_xticks(x)
ax.set_xticklabels(metrics)
ax.legend()

# Adding value annotations on the bars
for bar in bars1 + bars2:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width() / 2, yval + 0.2, f'{yval:.2f}', ha='center', va='bottom', fontsize=10)

# Minimalist design tweaks
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.yaxis.grid(True, linestyle='--', alpha=0.7)
ax.set_axisbelow(True)

# Show the plot
plt.tight_layout()
plt.show()

# Save the plot as a PDF
plt.savefig("barchart.pdf", format="pdf")
```
