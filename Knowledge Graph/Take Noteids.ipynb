```python
import pandas as pd

# Define the bucket and file names
bucket_name = 'mimicivliza'  # Replace with your bucket name
diagnoses_icd = f's3://{bucket_name}/diagnoses_icd.csv'
dict_diagnosis_icd = f's3://{bucket_name}/d_icd_diagnoses.csv'
dict_procedures_icd = f's3://{bucket_name}/d_icd_procedures.csv'
procedures_icd = f's3://{bucket_name}/procedures_icd.csv'
prescriptions= f's3://{bucket_name}/prescriptions.csv'

# Load the files
diagnoses_icd = pd.read_csv(diagnoses_icd)
procedures_icd = pd.read_csv(procedures_icd)
prescriptions_icd = pd.read_csv(prescriptions)
dict_diagnosis_icd = pd.read_csv(dict_diagnosis_icd)
dict_procedures_icd = pd.read_csv(dict_procedures_icd)
```

```python
diagnoses_icd.head(5)
```

```python
# Keep only the subject_id and icd_version columns
diagnoses_icd = diagnoses_icd[['subject_id', 'icd_version', 'icd_code']]

# Display the resulting dataframe
diagnoses_icd.head(5)
```

```python
# Merge the dataframes on icd_code and icd_version
merged_df_diagnosis = diagnoses_icd.merge(dict_diagnosis_icd, on=["icd_code", "icd_version"], how="left")

merged_df_diagnosis.head(5)
```

```python
# Merge the dataframes on icd_code and icd_version
merged_df_procedures = procedures_icd.merge(dict_procedures_icd, on=["icd_code", "icd_version"], how="left")

merged_df_procedures.head(5)
```

```python
# Keep only the subject_id and icd_version columns
merged_df_procedures = merged_df_procedures[['subject_id', 'icd_version', 'icd_code', 'long_title']]
```

```python
prescriptions_icd = prescriptions_icd[['subject_id', 'drug', 'prod_strength']]
```

```python
prescriptions_icd.head(5)
```

```python
# Finding common subject_id across the three dataframes
common_subject_ids = set(merged_df_procedures['subject_id']) & \
                     set(merged_df_diagnosis['subject_id']) & \
                     set(prescriptions_icd['subject_id'])

# Filtering dataframes to include only rows with common subject_id
common_df_procedures = merged_df_procedures[merged_df_procedures['subject_id'].isin(common_subject_ids)]
common_df_diagnosis = merged_df_diagnosis[merged_df_diagnosis['subject_id'].isin(common_subject_ids)]
common_df_prescriptions = prescriptions_icd[prescriptions_icd['subject_id'].isin(common_subject_ids)]

# Print the common subject IDs
print("Common Subject IDs:", common_subject_ids)

```

```python
# Count the unique common subject IDs
unique_common_subject_ids_count = len(common_subject_ids)

# Print the count
print("Number of unique common Subject IDs:", unique_common_subject_ids_count)
```

```python
# Merging the filtered dataframes on the common subject IDs
combined_df = (
    common_df_procedures
    .merge(common_df_diagnosis, on="subject_id", how="outer", suffixes=('_procedures', '_diagnosis'))
    .merge(common_df_prescriptions, on="subject_id", how="outer", suffixes=('', '_prescriptions'))
)

# Display the resulting dataframe
print("Combined DataFrame with unique common Subject IDs and all columns:")

```

```python
len(combined_df)
```

```python
combined_df.head(10)
```

```python
grouped_combined_df = combined_df.groupby('subject_id').agg({
    'icd_code_procedures': lambda x: list(x.unique()),    # List of unique procedure codes
    'long_title_procedures': lambda x: list(x.unique()),  # List of unique procedure descriptions
    'icd_code_diagnosis': lambda x: list(x.unique()),     # List of unique diagnosis codes
    'long_title_diagnosis': lambda x: list(x.unique()),   # List of unique diagnosis descriptions
    'drug': lambda x: ', '.join(map(str, x.unique())),    # Convert to strings and concatenate drugs
    'prod_strength': lambda x: ', '.join(map(str, x.unique()))  # Convert to strings and concatenate strengths
}).reset_index()

# Display the result
grouped_combined_df.head(5)
```

```python
len(grouped_combined_df)
```

```python
# Save the grouped dataframe to a CSV file
grouped_combined_df.to_csv('grouped_combined_df.csv', index=False)

print("The dataframe has been saved as 'grouped_combined_df.csv'.")
```
