```python
from tqdm import tqdm
from transformers import pipeline, AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig
import torch
import os

# Set environment variable for better memory management
os.environ["PYTORCH_CUDA_ALLOC_CONF"] = "expandable_segments:True"

# Configure quantization (8-bit)
quantization_config = BitsAndBytesConfig(
    load_in_8bit=True,
    llm_int8_enable_fp32_cpu_offload=True
)

# Load the model and tokenizer
model_name = "meta-llama/Llama-3.2-1B-Instruct"
model = AutoModelForCausalLM.from_pretrained(
    model_name,
    quantization_config=quantization_config,
    device_map="auto"
)

tokenizer = AutoTokenizer.from_pretrained(model_name)
# Fix the padding warning
tokenizer.padding_side = 'left'
tokenizer.pad_token_id = tokenizer.eos_token_id

# Initialize the pipeline
summarizer = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer
)

batch_size = 8
max_new_tokens = 500

# Prepare inputs
inputs = sample_data["input"].tolist()
generated_summaries = []

# Modified process_batch function to handle outputs correctly
def process_batch(batch):
    try:
        outputs = summarizer(
            batch,
            max_new_tokens=max_new_tokens,
            batch_size=batch_size,
            do_sample=True
        )
        # Fix the output processing
        return outputs if isinstance(outputs, list) else [outputs]
        
    except RuntimeError as e:
        if "out of memory" in str(e):
            torch.cuda.empty_cache()
            print(f"OOM error, retrying with batch size {len(batch)//2}")
            if len(batch) > 1:
                half = len(batch)//2
                return (
                    process_batch(batch[:half]) +
                    process_batch(batch[half:])
                )
        raise e

# Process with progress bar
with tqdm(total=len(inputs), desc="Processing Rows", unit="row") as pbar:
    for i in range(0, len(inputs), batch_size):
        batch = inputs[i:i + batch_size]
        
        # Process batch
        summaries = process_batch(batch)
        # Extract generated text correctly
        generated_texts = [summary[0]['generated_text'] if isinstance(summary, list) else summary['generated_text'] for summary in summaries]
        generated_summaries.extend(generated_texts)
        
        # Clear cache
        torch.cuda.empty_cache()
        
        # Update progress
        pbar.update(len(batch))

# Add to DataFrame and save
sample_data["generated_summary"] = generated_summaries
sample_data.to_csv("generated_summaries.csv", index=False)
print("Summaries saved to 'generated_summaries.csv'")
```
